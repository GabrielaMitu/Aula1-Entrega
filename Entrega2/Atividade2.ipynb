## Parte 1 - calibração

Ouça a explicação do professor sobre o modelo de câmera *pinhole*  e desenhe a medida $f$ que separa o plano focal da pupila da câmera

**Detalhe como calculou $f$:**

Foi utilizada a semelhança de triângulos: f/h = D/H 

Legenda: 

- f --> distância entre o plano focal e a pupila da câmera (em pixels)
- h --> altura do objeto na imagem (em pixels)
- D --> distância do objeto até a pupila da câmera (em centímetros)
- H --> altura do objeto real (em centímetros)

No experimento, foi medido:

H = 14cm, 
h = 440px, 
D = 22.5cm

Então, utilizando a fórmula, **f = 707px**
